```python
from bs4 import BeautifulSoup
import requests
import urllib.request
```

```python
url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
page=requests.get(url)

soup=BeautifulSoup(page.content , "html.parser")
```

```python
names=soup.find_all("td", class_="titleColumn")
Title=[]
Director=[]
Year=[]
for name in names:
    Title.append(name.find("a").get_text())
    Director.append(name.find("a")["title"])
    Year.append(name.find("span", class_="secondaryInfo").get_text().replace("(","").replace(")",""))
```

```python
links=soup.find_all("td",class_="posterColumn")
Link=[]
for i in links:
    Link.append("https://www.imdb.com/"+i.find("a")["href"])
    
```

```python
rates=soup.find_all("td", class_="imdbRating")
Rating=[]
for j in rates:
    Rating.append(j.find("strong").get_text())
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"TITLE":Title,"YEAR":Year, "DIRECTOR":Director,"LINKS":Link,"RATING":Rating})
```

```python
df
```

```python
df[["DIRECTOR1","ACTOR_1","ACTOR_2"]]=df.DIRECTOR.str.split(',',expand=True)
```
